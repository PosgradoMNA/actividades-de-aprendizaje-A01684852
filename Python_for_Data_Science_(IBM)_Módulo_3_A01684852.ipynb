## for loops
Ejecuta una actividad una y otra vez


Por ejemplo si queremos reemplazar cada uno de los cuadrados por un cuadrado blanco tendríamos que pedir uno por uno 

image.png

En python representaremos esto con una lista en donde cada elemento es un string representando un color, queremos cambiar el nombre, usamos el siguiente código en donde el valor de i se incrementa automáticamente en 1 cada vez, y el range indica cuántas veces se realizará esta función.

El valor de i es cero y cada iteración i aumenta en 1



```python
squares=["red", "yellow", "green", "purple", "blue"]

print(squares)

for i in range(0,5):
  squares[i]="white"

print(squares)
```

También podemos iterar a través de una lista o tuple de manera directa, sin necesidad de utilizar índices, esta función itera en todos los elementos de la lista

```python
squares=["red", "yellow", "green", "purple", "blue"]
for square in squares:
  square
```

otra forma

```python
squares=["red", "yellow", "green", "purple", "blue"]
for i, square in enumerate(squares):
  square
  print(i)
  print(square)
```

## While loops
Solo corre mientras se cumpla la condición, a diferencia del for, en while si debemos poner una variable que vaya sumando, en este código con la función append vamos guardando esos datos

```python
squares=['orange','orange','purple','blue']
Newsquares=[]
i=0

while(squares[i]=='orange'):
  Newsquares.append(squares[i])
  i=i+1
print(Newsquares)
```

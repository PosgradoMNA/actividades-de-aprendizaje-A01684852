## Scientifics Computing Libraries in Python

Una librería es una colección de funciones y métodos que permiten hacer muchísimas acciones sin escribir ningún código, normalmente las librerías contienen built-in modules, que proveen distintas funcionalidades que podemos utilizar directamente. Existen muchas librería ofreciendo muchas soluciones

Podemos dividir en 3 grupos:

1.- Scientific Computing Libraries: 
  - Pandas: Ofrece accesi rápido a datos estructurados, el principal instrumento de Pandas es una tabla bi dimensional que sonsiste en filas y columnas que son llamadas DataFrame, está designado para proveer una funcionalidad de indexación sencilla
  - Numpy: Usa arrays para sus inputs y outputs, puede ser extendido a objetos y matrices, y con cambios menores, los desarrolladores pueden desarrollar procesos rápidos en arrays
  - SciPy: Incluye funciones para algunos problemas matemáticos avanzados y visualización de datos


2.- Visualización: Usando métodos de visualización de datos, es la mejor forma para comunicar con otros mostrando los resultados de los análisis, estas librerías permiten crear gráficas, charts, y mapas
  - Matplotlib: Es la librería más conocida para visualización de datos, es muy buena para hacer gráficas y plots, las gráficas son altamente costumizables
  - Seaborn: Está basada en Matplotlib, es muy sencillo generar distintas gráficas como mapas de calor, series de tiempo y gráficas de violin


3.- Machine Learning Algoritm: podemos generar modelos con nuestros datasets y obtener predicciones
  - Scikit-learn: contiene herramientas para modelos estadísticos, incluyendo regresiones, clasificaciones, clustering y más, esta librería está construída sobre NumPy, SciPy y Matplotlib
    - StatsModels: Permite a los usuarios explorar datos, estimar modelos estadísticos y realizar pruebas estadísticas

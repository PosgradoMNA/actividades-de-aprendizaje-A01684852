# Reducción de Datos

Dado que la minería de datos es una técnica que se utiliza para manejar grandes cantidades de datos. Mientras trabajaba con un gran volumen de datos, el análisis se volvió más difícil en tales casos. Para deshacernos de esto, utilizamos la técnica de reducción de datos. Su objetivo es aumentar la eficiencia del almacenamiento, reducir los costos de análisis y almacenamiento de datos.

La reducción de datos es una técnica que se puede aplicar para obtener una representación reducida de un conjunto de datos que es mucho más pequeño en volumen en comparación con el conjunto de datos original pero que aún contiene información crítica. Presentamos a continuación algunas estrategias de reducción de datos:

 - Muestreo: Selección de un subconjunto de los objetos de datos a analizar.
 - Selección de características: selección de un subconjunto de las características que se van a analizar.
 - Reducción de la dimensionalidad: Crear nuevas funciones que sean una combinación de las características antiguas.

## Muestreo

El muestreo es un enfoque comúnmente utilizado para seleccionar un subconjunto de los datos que se van a analizar. Por lo general, se usa porque es demasiado costoso o requiere mucho tiempo para procesar todos los datos.

La idea clave en el muestreo es obtener una muestra representativa de los datos.

Algunas técnicas de muestreo incluyen lo siguiente:

 - Muestreo sistemático: Selección de instancias de una ventana de muestreo ordenada.
 - Muestreo aleatorio simple: Toda muestra posible del mismo tamaño tiene la misma probabilidad de ser seleccionada.
 - Muestreo aleatorio estratificado: Dividir los datos en grupos (estratos) y seleccionar una muestra aleatoria de cada grupo.
 - Muestreo por Clusters: Dividir los datos en grupos (conglomerados/Clusters) y seleccionar todos los miembros de uno o más, pero no todos, de los clusters.

## Selección de características

Para la selección de características, seleccionamos un conjunto mínimo de características tal que la distribución de probabilidad de la clase sea cercana a la obtenida por todas las características.

- Un buen vector de características se define por su capacidad para discriminar entre ejemplos de diferentes clases.

- Por ejemplo, maximizar la separación entre clases y minimizar la separación dentro de clases.

## Principal Component Analysis (PCA)

El análisis de componentes principales (PCA) es un procedimiento estadístico que utiliza una transformación ortogonal para convertir un conjunto de observaciones, de variables posiblemente correlacionadas, en un conjunto de valores de variables linealmente no correlacionadas denominadas componentes principales.

El número de componentes principales distintos es igual al menor número de variables originales o al número de observaciones menos uno. Esta transformación se define de tal manera que el primer componente principal tiene la mayor varianza posible (es decir, representa la mayor cantidad posible de variabilidad en los datos), y cada componente subsiguiente, a su vez, tiene la mayor varianza posible bajo la restricción de que es ortogonal a los componentes precedentes.

Los vectores resultantes son un conjunto de bases ortogonales no correlacionadas.

## Principal component regression (PCR) 

PCR funciona en tres pasos:

- Aplique PCA para generar componentes principales a partir de las variables predictoras, con el número de componentes principales que coincida con el número de características originales p
- Mantenga los primeros k componentes principales que explican la mayor parte de la varianza (donde k < p), donde k se determina mediante validación cruzada
- Ajuste un modelo de regresión lineal (usando mínimos cuadrados ordinarios) en estos k componentes principales

La idea es que el menor número de componentes principales represente la mayor parte de la variabilidad en los datos y (presuntamente) la relación con la variable objetivo. Por lo tanto, en lugar de utilizar todas las funciones originales para la regresión, solo utilizamos un subconjunto de los componentes principales.

Aunque la suposición de una relación con la variable objetivo no siempre se cumple, a menudo es una aproximación lo suficientemente razonable como para producir buenos resultados. 

**Ventajas**

- PCR ajusta un modelo de regresión lineal en k componentes principales en lugar de todas las características originales, lo que ayuda a reducir el sobre-ajuste.
- PCR ayuda a eliminar la multi-colinealidad en los datos al eliminar los componentes principales asociados con valores propios pequeños.
- La mejora del rendimiento es más destacada en datos con muchas características (alta cardinalidad) y multi-colinealidad significativa.

**Desventajas**

- PCR no se considera un método de selección de características porque los componentes principales resultantes utilizados en la regresión son combinaciones lineales de las características originales. Significa que seguimos confiando en todas las funciones originales en lugar de en un subconjunto de ellas.
- Dado que los componentes principales se utilizan en el modelo, estos predictores pierden su "explicabilidad" en comparación con las características originales.
- La transformación PCA no está supervisada, lo que significa que no considera la variable objetivo al determinar los componentes principales. Por lo tanto, no hay garantía de que los componentes principales con la mayor varianza sean los mejores para predecir el objetivo.
 

##Partial least squares regression (PLSR)

Cuando el algoritmo de aprendizaje es una regresión lineal, el modelo resultante se conoce como regresión de componentes principales.

Dado que los componentes principales son en sí mismos combinaciones lineales de los atributos originales, el resultado de la regresión de componentes principales se puede volver a expresar en términos de los atributos originales.

De hecho, si se utilizan todos los componentes, no solo los “principales”, el resultado es el mismo que se obtiene al aplicar la regresión de mínimos cuadrados a los datos de entrada originales.

- Estandarice las variables predictoras y de respuesta.
- Calcule M combinaciones lineales (llamadas "componentes PLS") de las p variables predictoras originales que explican una cantidad significativa de variación tanto en la variable de respuesta como en las variables predictoras.
- Utilice el método de mínimos cuadrados para ajustar un modelo de regresión lineal utilizando los componentes de PLS como predictores.
- Utilice la validación cruzada k-fold para encontrar el número óptimo de componentes de PLS para mantener en el modelo.

##Multidimensional scaling analysis (MDS)

El escalado multidimensional (MDS) es una técnica que crea un mapa que muestra las posiciones relativas de una serie de objetos, dada solo una tabla de las distancias entre ellos.

El mapa puede constar de una, dos, tres o incluso más dimensiones. El programa calcula la solución métrica o no métrica. La tabla de distancias se conoce como matriz de proximidad.

Surge directamente de los experimentos o indirectamente como una matriz de correlación.

Dentro del Escalamiento Multidimensional, existen métodos para diferentes tipos de datos:

- El escalamiento multidimensional métrico (Análisis de coordenadas principales): Trata con distancias numéricas, en las que no hay error de medición (tiene exactamente una medida de distancia para cada par de elementos).
- El escalamiento multidimensional no métrico: Se ocupa de distancias no numéricas entre elementos, en las que no hay error de medición (tiene exactamente una medida de distancia para cada par de elementos).
- El escalamiento de diferencias individuales: Se aplica cuando tiene varias estimaciones (diferentes) de las distancias entre los elementos. Este suele ser el caso cuando varios individuos dan cada uno una estimación de las distancias entre todos los pares.
- El análisis multidimensional de preferencia: Se aplica cuando tiene varias estimaciones (diferentes) de las distancias entre los elementos. Este método no funciona en matrices de distancia, sino que funciona en la clasificación de datos (múltiples participantes clasifican cada elemento de mejor a peor).

##Projection pursuit method (PP)

Es una técnica para el análisis exploratorio de conjuntos de datos multivariados. Este método busca proyecciones lineales interesantes de los datos multivariados en una línea o plano.

##Linear Discriminant Analysis (LDA)

Se utiliza para modelar diferencias en grupos, es decir, separar dos o más clases. Se utiliza para proyectar las características en un espacio de mayor dimensión en un espacio de menor dimensión

##Mixed Discriminant Analysis (MDA)

Para MDA, hay clases, y se supone que cada clase es una mezcla gaussiana de subclases, donde cada punto de datos tiene una probabilidad de pertenecer a cada clase. Aún se asume la igualdad de la matriz de covarianza, entre clases.

##Quadratic Discriminant Analysis (QDA)

Al igual que LDA, el clasificador QDA resulta de asumir que las observaciones de cada clase se extraen de una distribución gaussiana y de conectar las estimaciones de los parámetros en el teorema de Bayes para realizar la predicción.

En QDA debemos asumir que la matriz de covarianza puede ser diferente para cada clase y, por lo tanto, es estimar la matriz de covarianza por separado para cada clase

## File Object
Para leer información de un archivo utilizamos la función .read()
Por ejemplo:

File1 = open("/resources/data/Example2.txt","w")

Viene la función open, después el link del archivo, después nombre del archivo y al final la función, si usamos "r" es read, "w" es write y "a" for appending

Si ya tenemos el File1, después podemos buscar el nombre del archivo a través de:

File1.name

Y ver si solo es lectura o escritura con 
File1.mode()

Lo mejor es abrir el arcivo con el statement with, ya que con esto nosotros cerramos el archivo una vez copiadala información, y todo el código se corre en el bloque



```python
with open("Example1.txt","r") as File1: # no va a correr el código pero se pone el ejemplo

  file_stuff=File1.read() # guarda una copia del contenido en la variable file_stuff
  print(file_stuff) #imprime el contenido

print(File1.closed) #revisa si se cerró el archivo
print(file_stuff) #se imprimer el contenido desde el bloque
```

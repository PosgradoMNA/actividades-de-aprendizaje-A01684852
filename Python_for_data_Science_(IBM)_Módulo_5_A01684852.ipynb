#### Vector Addition and Substraction

al sumar dos vectores se genera la suma de cada uno de sus elementos

image.png

Si analizamos esto en vectores es más sencillo entenderlo

image.png

```python
u=np.array([1,0])
v=np.array([0,1])
z=u+v
z
```

```python
# Vector addition con una línea de código numpy, el código correrá más rápido
u=np.array([1,0])
v=np.array([0,1])
z=[]

for n, m in zip(u,v):
  z.append(n+m)
z
```

```python
u=np.array([1,0])
v=np.array([0,1])
z=u-v
z
```

```python
u=np.array([1,0])
v=np.array([0,1])
z=[]

for n, m in zip(u,v):
  z.append(n-m)
z
```

```python
import numpy as np
from sklearn.decomposition import PCA

pcs = PCA(n_components=2)

pcs.fit(df[['ToPay_Sep', 'ToPay_Aug']])

pcsSummary = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                          'Proportion of variance': pcs.explained_variance_ratio_,
                          'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)#column 
                          }
                          )
pcsSummary
```

El primer componente tiene una varianza de 97 
No quiere decir que ToPay_Sep, es 97, lo que quiere decir es que el componente de ToPay_Sep y ToPay_Aug es 97, entonces, podemos eliminar los dos anteriores y solo dejar este componente

```python
pcsSummary = pcsSummary.transpose()
pcsSummary.round(2)
```

Renombramos como Componente PC1 y PC2

```python
pcsSummary.columns = ['PC1', 'PC2']
pcsSummary = pcsSummary.round(4)
pcsSummary
```

z1 cuenta por el 97.58% del total de la variabilidad
z2 por el restante 2.42%

Por lo tanto si eliminarmos z2 = PC2, seguimos manteniento el 97.58% del total de variabilidad

## Scree Plot

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```

Lo anterior explica el componente de varianza, y la línea el valor acumulado

```python
pcs.components_.round(4) #is the set of all eigenvectors for your projection space (one eigenvector for each principal component).
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), 
                                columns=['PC1', 'PC2'],
                                index=['ToPay_Sep', 'ToPay_Aug']
                                )
pcsComponents_df
```

El principal componente 1, tiene un impacto de 0.71, el segundo de 0.69
